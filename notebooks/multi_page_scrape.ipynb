```python
import requests
from bs4 import BeautifulSoup as bs
from IPython.core.display import HTML
import pandas as pd

URL = 'https://ryman.com/events/'
response = requests.get(URL)

soup = bs(response.text)
```

```python
h6 = soup.findAll('p')
```

```python
page = 1
artistsmulti = []
while page != 6:
      url = f"https://ryman.com/events/list/?tribe_event_display=list&tribe_paged={page}"
      response = requests.get(url)
      html = response.content
      soup = bs(html, "lxml")
      for h3 in soup.find_all('a', attrs={'class': 'tribe-event-url'}):
            artistsmulti.append(h3.get_text(strip=True))
      page = page + 1
artistsmulti
```

```python
page = 1
tdmulti = []
while page != 6:
      url = f"https://ryman.com/events/list/?tribe_event_display=list&tribe_paged={page}"
      response = requests.get(url)
      html = response.content
      soup = bs(html, "lxml")
      for h3 in soup.find_all('div', attrs={'class': 'tribe-event-schedule-details'}):
            tdmulti.append(h3.get_text())
      page = page + 1
tdmulti
```
