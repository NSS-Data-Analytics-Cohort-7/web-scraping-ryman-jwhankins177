```python
import requests
from bs4 import BeautifulSoup as bs
from IPython.core.display import HTML
import pandas as pd
```

List of artists, page 1 - 5.

```python
page = 1
artistsmulti = []
while page != 6:
      url = f"https://ryman.com/events/list/?tribe_event_display=list&tribe_paged={page}"
      response = requests.get(url)
      html = response.content
      soup = bs(html, "lxml")
      for h3 in soup.find_all('a', attrs={'class': 'tribe-event-url'}):
            artistsmulti.append(h3.get_text(strip=True))
      page = page + 1
artistsmulti
```

Data and time for all shows, page 1 - 5.

```python
page = 1
tdmulti = []
while page != 6:
      url = f"https://ryman.com/events/list/?tribe_event_display=list&tribe_paged={page}"
      response = requests.get(url)
      html = response.content
      soup = bs(html, "lxml")
      for h3 in soup.find_all('div', attrs={'class': 'tribe-event-schedule-details'}):
            tdmulti.append(h3.get_text())
      page = page + 1
tdmulti
```

Clean and separate dates and times. 

```python
# Cleaned up dates
datemulti = [x.split(', ',1)[-1] for x in tdmulti]
datemulti = [x.split(' at ',1)[0] for x in datemulti]
datemulti 
```

```python
# Cleaned up time
timemulti = [x.split(' at ',1)[-1] for x in tdmulti]
timemulti = [x.split('\n' ,1)[0] for x in timemulti]
timemulti
```

Make tables DataFrames.

```python
artistsmulti = pd.DataFrame(artistsmulti)
datemulti = pd.DataFrame(datemulti)
timemulti = pd.DataFrame(timemulti)
```

Combine three tables into one. Rename columns.

```python
schedule_page_1_5 = pd.concat([artistsmulti,datemulti,timemulti], axis = 1)
schedule_page_1_5.columns = ['artist','date', 'time']
schedule_page_1_5
```

```python
page = 1
opensmulti = []
while page != 6:
      url = f"https://ryman.com/events/list/?tribe_event_display=list&tribe_paged={page}"
      response = requests.get(url)
      html = response.content
      soup = bs(html, "lxml")
      for h3 in soup.find_all('span', attrs={'class': 'opener'}):
            if h3 == h3:
                opensmulti.append(h3.get_text(strip=True))
            if h3 != h3:
                opensmulti.append('No opening act.')
      page = page + 1
opensmulti = pd.DataFrame(opensmulti)
```
